```python
import pandas as pd
```

```python
df = pd.read_csv("/home/crdamke/Documentos/Python/app/app_falhas/dados_unificados.csv")
df.head()
```

```python
df["timestamp"] = pd.to_datetime(df["timestamp"], errors="coerce")
```

```python
df.head()
```

```python
df = df.sort_values("timestamp").reset_index(drop=True)
df.head()
```

```python
df_copy = df.copy()
df_copy.head()
```

```python
df_copy['delta'] = df_copy['timestamp'].diff().dt.total_seconds().fillna(0)
df_copy.head()
```

```python
linhas_vazias = df_copy[df_copy['delta'] > 10]
linhas_vazias
```

```python
intervalos_falhas = []
tempo_inicio = df_copy['timestamp'].min()
tempo_inicio
```

```python
# Calcular a diferença em horas desde o início até cada linha com falha
intervalos_desde_inicio = (linhas_vazias['timestamp'] - tempo_inicio).dt.total_seconds() / 3600
intervalos = intervalos_desde_inicio.tolist()
intervalos
```

```python
# Cria um DataFrame vazio para armazenar os intervalos
df_intervalos = pd.DataFrame()
```

```python
# Adiciona a coluna 'timestamp' ao DataFrame de intervalos
df_intervalos['timestamp'] = df['timestamp']
df_intervalos.head()
```

```python
df_intervalos['temp'] = df['temp']
df_intervalos.head()
```

```python
# Calcula a diferença de tempo entre timestamps consecutivos em segundos
df_intervalos['delta'] = df_intervalos['timestamp'].diff().dt.total_seconds().fillna(0)
df_intervalos.head()
```

```python
num_linhas = len(df_intervalos)
num_linhas2 = len(df_intervalos['delta'])
num_linhas, num_linhas2
```

```python
for i in range(num_linhas):
    if df_intervalos['delta'][i] > 10:
        print(f"Intervalo maior que 10 segundos encontrado na linha {i}, delta: {df_intervalos.loc[i, 'delta']} segundos")
```
