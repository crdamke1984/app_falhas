```python
def teste_type(tipo):
    if tipo is list:
        return "É uma lista"
    elif tipo is str:
        return "É uma string"
    elif tipo is dict:
        return "É um dicionário"
    else:
        return "Tipo desconhecido"
```

```python
sensor = "A"
```

```python
teste_type(type(sensor))
```

```python
def dict_urls(sensor: list, url: list):
    files_url = {}
    for i in range(len(sensor)):
        files_url[sensor[i]] = url[i]
    return files_url
```

```python
sensor = ["A"]
url = ["url_A"]
```

```python
files = dict_urls(sensor, url)
print(files)
```

```python
import requests

def request_data(dict_urls: dict):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
        }
    for key, url in dict_urls.items():
        response = requests.get(url, headers=headers)
        if response.status_code == 200:
            dados = response.content.decode('utf-8')
        else:
            print(f"Erro ao acessar {url}: {response.status_code}")
    return dados
```

```python
sensor = ['energia']
url = ['http://crdamke.cloud/extrator']
```

```python
from io import StringIO

files = dict_urls(sensor, url)
dados = request_data(files)
csv_data = StringIO(dados)


print(type(csv_data))
```

```python
import pandas as pd
from io import StringIO

csv_data = StringIO(dados)

df = pd.read_csv(csv_data, sep=None)
print(df.head())
```
